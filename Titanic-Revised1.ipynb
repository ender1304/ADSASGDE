```python
plot=data_Train.groupby('Pclass').Survived.mean().plot(kind='bar')
plot.set_title("Survived by class")
plot.set_xlabel("1st, 2nd or 3rd class")
plot.set_ylabel("mean survival")
fig=plot.get_figure()
fig.savefig("c:\\Kaggle\\Survived by class.png")
```

```python

data_Train['AgeRange']=pd.cut(data_Train['Age'],[0,15,30,45],labels=['0-15','15-30','30-45'])
clean_age=data_Train.dropna(subset=['Age'])
plot=clean_age.groupby('AgeRange')['Survived'].mean().plot(kind='bar')
plot.set_title("Survived by Age Range")
plot.set_xlabel("years of age")
plot.set_ylabel("mean survival")
fig=plot.get_figure()
fig.savefig("c:\\Kaggle\\Survived by age range.png")
```

```python
plot=data_Train.groupby('Sex').Survived.mean().plot(kind='bar')
plot.set_title("Survived by Gender")
plot.set_xlabel("male or female")
plot.set_ylabel("mean survival")
fig=plot.get_figure()
fig.savefig("c:\\Kaggle\\Survived by gender.png")
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
dtree = DecisionTreeClassifier()
```

```python
X_train=data_Train[['Pclass']]
y = data_Train['Survived']
X_test=data_Test[['Pclass']]
```

```python
passengerid=data_Test['PassengerId']
```

```python
#make prediction on remaining 1/3 of data set
dtree.fit(X_train,y)
prediction=dtree.predict(X_test)
data_prediction=pd.DataFrame(data=np.column_stack(
        [passengerid,prediction]),index=data_Test.index.values,
        columns=['PassengerId','Survived'])
```

```python
print(data_prediction)
data_prediction.info()
```

```python
#check format of prediction, save to file
data_prediction.to_csv('c:\\kaggle\\myprediction.csv',index=False)
```
